```python
import pandas as pd
import numpy as np
```

```python
### only works for 2019 to 2017 since they were already in excel format
files= ["2019 Annual Insurance Industry Statistics.xlsm","2018 Annual Statistics 28062019.xls","2017-Annual-Insurance-Statistics-Uploaded.xlsm"]
new_df= pd.DataFrame()
c=0
for fil in files:
    c+=1
    f1 = pd.ExcelFile(fil)
    ## reading the specific sheets
    sheet1= pd.read_excel(f1,"APPENDIX 4 I",header=3)
    sheet2 = pd.read_excel(f1,"APPENDIX 4 II",header=3)
    sheet3 = pd.read_excel(f1,"APPENDIX 4 III",header=4)
    ### dropping the unnamed column
    sheet1.drop("Unnamed: 0",axis=1,inplace=True)
    sheet2.drop("Unnamed: 0",axis=1,inplace=True)
    sheet3.drop("Unnamed: 0",axis=1,inplace=True)
    ### setting the df index
    sheet1.set_index("Company",inplace=True)
    sheet2.set_index("Company",inplace=True)
    sheet3.set_index("Company",inplace=True)

    ### concatinating the data sheets
    new_df = pd.concat([new_df,sheet1.T,sheet2.T,sheet3.T])





    ### checing number of rows to confirm match
# print(globals()["sheet1_"+str(c)].shape[0])
# print(globals()["sheet2_"+str(c)].shape[0])
# print(globals()["sheet3_"+str(c)].shape[0])

```
